```python
import os
import pandas as pd
harmonies = []
path = "./harmonies" #文件夹目录
files= os.listdir(path) 
for file in files:
    if not os.path.isdir(file):
        train=pd.read_csv(path+"/"+file, sep='\t')
        harmonies.append(train)
```

```python
harmonies[0]
```

```python
chord = []
c_type = []
for i in harmonies:
    chord = chord + list(i['chord'])
    c_type = c_type + list(i['chord_type'])
```

```python
def Count(List):
    dict={}
    for area in List:
        keys=area.split(",")
        for key in keys:
            if key in dict.keys():
                dict[key]=dict[key]+1
            else:
                dict[key]=1
    return dict
```

```python
count_chord = Count(chord)
count_ctype = Count(c_type)
```
