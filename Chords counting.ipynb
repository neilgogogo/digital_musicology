```python
import os
import pandas as pd
harmonies = []
path = "./harmonies" #文件夹目录
files= os.listdir(path) 
for file in files:
    if not os.path.isdir(file):
        train=pd.read_csv(path+"/"+file, sep='\t')
        harmonies.append(train)
```

```python
harmonies[0]
```

```python
chord = []
c_type = []
for i in harmonies:
    chord = chord + list(i['chord'])
    c_type = c_type + list(i['chord_type'])
```

```python
chord[0:10]
```

```python
c_type[0:10]
```

```python
def Count(List):
    dict={}
    for area in List:
        keys=area.split(",")
        for key in keys:
            if key in dict.keys():
                dict[key]=dict[key]+1
            else:
                dict[key]=1
    return dict
```

```python
count_chord = Count(chord)
count_ctype = Count(c_type)
```

```python
count_chord
```

```python
count_ctype
```

```python
late_stage = ['127','133','131','132','133','135']
```

```python
late_harmony = []
for file in files:
    if not os.path.isdir(file):
        for i in late_stage:
            if i in file:
                train=pd.read_csv(path+"/"+file, sep='\t')
                late_harmony.append(train)              
```

```python
late_harmony[0]
```

```python
len(late_harmony)
```

```python
len(harmonies)
```

```python
late_chord = []
late_c_type = []
for i in late_harmony:
    late_chord = late_chord + list(i['chord'])
    late_c_type = late_c_type + list(i['chord_type'])
```

```python
late_count_chord = Count(late_chord)
late_count_ctype = Count(late_c_type)
```

```python
late_count_ctype
```

```python
sum(late_count_ctype.values())
```

```python
not_late_count_ctype = {}
for key in count_ctype.keys():
    not_late_count_ctype[key] = count_ctype[key] - late_count_ctype[key]

not_late_count_ctype
```

```python
sum(not_late_count_ctype.values())
```

```python
import matplotlib.pyplot as plt
width = 1.0
plt.bar(late_count_ctype.keys(), late_count_ctype.values(), width, color='g')
```

```python
plt.bar(not_late_count_ctype.keys(), not_late_count_ctype.values(), width, color='g')
```

```python
not_late_count_ctype_labels=list(sorted(not_late_count_ctype.keys()))
not_late_count_ctype_fracs=[not_late_count_ctype.get(not_late_count_ctype_labels[i]) for i in range(len(not_late_count_ctype_labels))]
fig=plt.figure()
plt.pie(x=not_late_count_ctype_fracs, labels=not_late_count_ctype_labels,autopct='%3.1f %%',shadow=True, labeldistance=1.1, startangle = 90,pctdistance = 0.6)
```

```python
late_count_ctype_labels=list(sorted(late_count_ctype.keys()))
late_count_ctype_fracs=[late_count_ctype.get(late_count_ctype_labels[i]) for i in range(len(late_count_ctype_labels))]
fig=plt.figure()
plt.pie(x=late_count_ctype_fracs, labels=late_count_ctype_labels,autopct='%3.1f %%',shadow=True, labeldistance=1.1, startangle = 90,pctdistance = 0.6)
```

We can see that the amount of chords in Beethoven's late stage works, which is widely accepted as work 127, 130, 131, 132, 133 and 135, is about half of the amount of chords in other stages'work (8726:16219). While some chords type proportion like M remains stable, we can see a increasing proportion of +, mm7 and MM7 (almost twice over that of the earlier stage' works), although they just make up of a small part of the chord.
